```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from PIL import Image
import os
from tqdm import tqdm,trange
import json
```

```python
out_dir="dataset/pannuke/Fold 1/images/fold1/"
arr=np.load(os.path.join(out_dir,"images.npy"))
```

```python
save_dir=os.path.join(out_dir,"separate_images")
os.makedirs(save_dir,exist_ok=True)
for i in trange(len(arr)):
    fp=os.path.join(save_dir,"%05d.png"%(i))
    Image.fromarray(arr[i].astype(np.uint8)).save(fp)
```

```python
pred_dir=os.path.join(out_dir,"separate_images_pred")
Image.open(os.path.join(pred_dir,"overlay","00003.png"))
```

```python
with open(os.path.join(pred_dir,"json","00000.json"),'r') as f:
    d=json.load(f)
```
